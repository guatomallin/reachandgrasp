```python
import matplotlib.pyplot as plt
import numpy as np

#making toy data
x = np.arange(0, 2, .05)

s = []
s.append(np.sin(x*3))
s.append(np.cos(x*4+0.3))
s.append(np.sin(x)*np.exp(2-x)*-1)
plt.clf()
plt.plot(x, s[0], x, s[1], x, s[2])
plt.show()
plt.clf()
```

```python
print(type(s))
```

```python
mindexes = [i.argmin() for i in s]
print(mindexes)#finding the minima
shifts = mindexes - min(mindexes)
print(shifts)
def shifted(shift, x, s):
    if shift == 0:   #x[:-0] was not what I wanted.
        return x, s
    else:
        return x[:-1*shift], s[shift:]

for i in (0,1,2):
    px, py = shifted(shifts[i], x, s[i])
    plt.plot(px, py)
plt.show()
```
